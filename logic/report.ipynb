### 語彙の定義

#### 定数記号
$Bluebeard$
: Bluebeard

$Charlie$
: Charlie
#### 述語
$Horse(x)$
: $x$は馬である

$Cow(x)$
: $x$は牛である

$Pig(x)$
: $x$は豚である

$Animal(x)$
: $x$は動物である

$Mammal(x)$
: $x$は哺乳動物である

$Parent(x,y)$
: $x$は$y$の親である

$Grandchild(x,y)$
: $x$は$y$の孫である

$Descendant(x,y)$
: $x$は$y$の子孫である

### a.
$$
\forall x(Horse(x)\lor Cow(x)\lor Pig(x)\Rightarrow Animal(x))
$$

#### 含意除去
$$
\forall x(\lnot(Horse(x)\lor Cow(x)\lor Pig(x))\lor Animal(x))
$$

#### ド・モルガンの法則
$$
\forall x((\lnot Horse(x)\land\lnot Cow(x)\land\lnot Pig(x))\lor Animal(x))
$$

#### 限量子除去
$$
(\lnot Horse(x)\land\lnot Cow(x)\land\lnot Pig(x))\lor Animal(x)
$$

#### $\land$の$\lor$上への分配
$$
(\lnot Horse(x)\lor Animal(x))\land(\lnot Cow(x)\lor Animal(x))\land(\lnot Pig(x)\lor Animal(x))
$$

#### 選言の含意への変換
$$
\left\{\begin{aligned}
Horse(x)&\Rightarrow Animal(x)\\
Cow(x)&\Rightarrow Animal(x)\\
Pig(x)&\Rightarrow Animal(x)
\end{aligned}\right.
$$

### b.
$$
\forall x(\exists y(Descendant(x,y)\land Horse(y))\Rightarrow Horse(x))
$$

#### 含意除去
$$
\forall x(\lnot\exists y(Descendant(x,y)\land Horse(y))\lor Horse(x))
$$

#### ド・モルガンの法則
$$
\begin{aligned}
&\forall x(\forall y\lnot(Descendant(x,y)\land Horse(y))\lor Horse(x))\\
\Leftrightarrow~&\forall x(\forall y(\lnot Descendant(x,y)\lor\lnot Horse(y))\lor Horse(x))
\end{aligned}
$$

#### 限量子の左移動
$$
\forall x\forall y((\lnot Descendant(x,y)\lor\lnot Horse(y))\lor Horse(x))
$$

#### 限量子除去
$$
(\lnot Descendant(x,y)\lor\lnot Horse(y))\lor Horse(x)
$$

#### 平滑化
$$
\lnot Descendant(x,y)\lor\lnot Horse(y)\lor Horse(x)
$$

#### 選言の含意への変換
$$
Descendant(x,y)\land Horse(y)\Rightarrow Horse(x)
$$

### c.
$$
Horse(Bluebeard)
$$

#### 含意への変換
$$
True\Rightarrow Horse(Bluebeard)
$$

### d.
$$
Parent(Bluebeard,Charlie)
$$

#### 含意への変換

True\Rightarrow Parent(Bluebeard,Charlie)
$$

### e.
$$
\forall x\forall y(Grandchild(x,y)\Leftrightarrow\exists z(Parent(z,x)\land Parent(y,z)))
$$

#### 含意除去
$$
\forall x\forall y((\lnot Grandchild(x,y)\lor\exists z(Parent(z,x)\land Parent(y,z)))\land(\lnot\exists z(Parent(z,x)\land Parent(y,z))\lor Grandchild(x,y)))
$$

#### ド・モルガンの法則
$$
\begin{aligned}
&\forall x\forall y((\lnot Grandchild(x,y)\lor\exists z(Parent(z,x)\land Parent(y,z)))\land(\forall z\lnot(Parent(z,x)\land Parent(y,z))\lor Grandchild(x,y)))\\
\Leftrightarrow~&\forall x\forall y((\lnot Grandchild(x,y)\lor\exists z(Parent(z,x)\land Parent(y,z)))\land(\forall z(\lnot Parent(z,x)\lor\lnot Parent(y,z))\lor Grandchild(x,y)))
\end{aligned}
$$

#### 変数標準化
$$
\forall x\forall y((\lnot Grandchild(x,y)\lor\exists w(Parent(w,x)\land Parent(y,w)))\land(\forall z(\lnot Parent(z,x)\lor\lnot Parent(y,z))\lor Grandchild(x,y)))
$$

#### 限量子の左移動
$$
\forall x\forall y\exists w\forall z((\lnot Grandchild(x,y)\lor(Parent(w,x)\land Parent(y,w)))\land((\lnot Parent(z,x)\lor\lnot Parent(y,z))\lor Grandchild(x,y)))
$$

#### スコーレム化
$$
\forall x\forall y\forall z((\lnot Grandchild(x,y)\lor(Parent(F(x,y),x)\land Parent(y,F(x,y))))\land((\lnot Parent(z,x)\lor\lnot Parent(y,z))\lor Grandchild(x,y)))
$$
$F(x,y)$: スコーレム関数

#### 全称限量子除去
$$
(\lnot Grandchild(x,y)\lor(Parent(F(x,y),x)\land Parent(y,F(x,y))))\land((\lnot Parent(z,x)\lor\lnot Parent(y,z))\lor Grandchild(x,y))
$$

#### $\land$の$\lor$上への分配
$$
((\lnot Grandchild(x,y)\lor Parent(F(x,y),x))\land(\lnot Grandchild(x,y)\lor Parent(y,F(x,y))))\land((\lnot Parent(z,x)\lor\lnot Parent(y,z))\lor Grandchild(x,y))
$$

#### 平滑化
$$
(\lnot Grandchild(x,y)\lor Parent(F(x,y),x))\land(\lnot Grandchild(x,y)\lor Parent(y,F(x,y)))\land(\lnot Parent(z,x)\lor\lnot Parent(y,z)\lor Grandchild(x,y))
$$

#### 選言の含意への変換
$$
\left\{\begin{aligned}
Grandchild(x,y)&\Rightarrow Parent(F(x,y),x)\\
Grandchild(x,y)&\Rightarrow Parent(y,F(x,y))\\
Parent(z,x)\land Parent(y,z)&\Rightarrow Grandchild(x,y)
\end{aligned}\right.
$$

### f.
$$
\forall x(Mammal(x)\Rightarrow\exists y Parent(y,x))
$$

#### 含意除去
$$
\forall x(\lnot Mammal(x)\lor\exists y Parent(y,x))
$$

#### 限量子の左移動
$$
\forall x\exists y(\lnot Mammal(x)\lor Parent(y,x))
$$

#### スコーレム化
$$
\forall x(\lnot Mammal(x)\lor Parent(G(x),x))
$$
$G(x)$: スコーレム関数

#### 全称限量子除去
$$
\lnot Mammal(x)\lor Parent(G(x),x)
$$

#### 選言の含意への変換
$$
Mammal(x)\Rightarrow Parent(G(x),x)
$$

## 2.

### 語彙の定義

#### 定数記号
$Taro$
: 太郎

$Tamatebako$
: 玉手箱

#### 述語
$BecomeOldPersonInstantly(x)$
: $x$は一瞬のうちに老人になる

$Open(x,y)$
: $x$は$y$を開ける

### 論理表現

#### 太郎は一瞬のうちに老人になった
$$
BecomeOldPersonInstantly(Taro)
$$

##### 含意への変換
$$
True\Rightarrow BecomeOldPersonInstantly(Taro)\tag{A}
$$

#### 玉手箱を開けると一瞬のうちに老人になる
$$
\forall x(Open(x,Tamatebako)\Rightarrow BecomeOldPersonInstantly(x))
$$

##### 含意除去
$$
\forall x(\lnot Open(x,Tamatebako)\lor BecomeOldPersonInstantly(x))
$$

##### 限量子除去
$$
\lnot Open(x,Tamatebako)\lor BecomeOldPersonInstantly(x)
$$

##### 選言の含意への変換
$$
Open(x,Tamatebako)\Rightarrow BecomeOldPersonInstantly(x)\tag{B}
$$

#### 太郎は玉手箱 をあけてしまったのだ
$$
Open(Taro,Tamatebako)
$$

##### 含意への変換
$$
True\Rightarrow Open(Taro,Tamatebako)\tag{C}
$$

### 最後の文を導出できるか議論
真であると導出することはできない．なぜなら，前の文章(A)と(B)を融合すると
$$
Open(Taro,Tamatebako)\Rightarrow BecomeOldPersonInstantly(Taro)\tag{D}
$$
にしかならず，(C)にはならないからである．

しかし，仮説として導出することはできる．(D)の前件文として，つまり「(C)が真であると仮定すると(B)と(C)の融合から(A)が真であると証明することができる」ということから(C)を導出することはできる．この種の推論は**アブダクション**（仮説生成）といい，結論から前提を（結果から原因を）探るものである．アブダクションは科学における重大な発見に辿り着くために重要であるだけでなく，診断を行うエキスパートシステム等の人工知能の分野でも研究されている[^abduction].

[^abduction]: [井上克巳. アブダクションの原理. 人工知能, 1992, 7.1: 48-59.](https://www.jstage.jst.go.jp/article/jjsai/7/1/7_48/_pdf)

## 3.

野獣の位置が問題であるから，穴については考えない．

### 語彙の定義

#### 定数記号
$W$
: 野獣

$[i,j]$
: マス($1\le i,j\le 4$)

#### 述語
$At(x,y)$
: $x$に$y$がある

$Ad(x,y)$
: $x$と$y$は隣接する

$S(x)$
: $x$で異臭がする

### 環境に関する知識

#### その部屋自身と隣接する部屋に野獣がいなければ異臭がしない
$$
\forall x(\lnot At(x,W)\land(\forall y(Ad(x,y)\Rightarrow\lnot At(y,W)))\Rightarrow\lnot S(x))
$$

##### 含意除去
$$
\forall x(\lnot(\lnot At(x,W)\land(\forall y(\lnot Ad(x,y)\lor\lnot At(y,W))))\lor\lnot S(x))
$$

##### ド・モルガンの法則
$$
\begin{aligned}
&\forall x(At(x,W)\lor\lnot(\forall y(\lnot Ad(x,y)\lor\lnot At(y,W)))\lor\lnot S(x))\\
\Leftrightarrow~&\forall x(At(x,W)\lor(\exists y\lnot(\lnot Ad(x,y)\lor\lnot At(y,W)))\lor\lnot S(x))\\
\Leftrightarrow~&\forall x(At(x,W)\lor(\exists y(Ad(x,y)\land At(y,W)))\lor\lnot S(x))
\end{aligned}
$$

##### 限量子の左移動
$$
\forall x\exists y(At(x,W)\lor(Ad(x,y)\land At(y,W))\lor\lnot S(x))
$$

##### スコーレム化
$$
\forall x(At(x,W)\lor(Ad(x,F(x))\land At(F(x),W))\lor\lnot S(x))
$$
$F(x)$: スコーレム関数

##### 全称限量子除去
$$
At(x,W)\lor(Ad(x,F(x))\land At(F(x),W))\lor\lnot S(x)
$$

##### $\land$の$\lor$への分配
$$
(At(x,W)\lor Ad(x,F(x))\lor\lnot S(x))\land(At(x,W)\lor At(F(x),W)\lor\lnot S(x))
$$

##### 選言の含意への変換
$$
S(x)\Rightarrow At(x,W)\lor Ad(x,F(x))\tag{A}
$$
$$
S(x)\Rightarrow At(x,W)\lor At(F(x),W)\tag{B}
$$

#### 異臭がしなければその部屋自身と隣接する部屋に野獣がいない
$$
\forall x(\lnot S(x)\Rightarrow\lnot At(x,W)\land(\forall y(Ad(x,y)\Rightarrow\lnot At(y,W))))
$$

##### 含意除去
$$
\forall x(S(x)\lor(\lnot At(x,W)\land(\forall y(\lnot Ad(x,y)\lor\lnot At(y,W)))))
$$

##### 全称限量子除去
$$
S(x)\lor(\lnot At(x,W)\land(\lnot Ad(x,y)\lor\lnot At(y,W)))
$$

##### $\land$の$\lor$への分配
$$
(S(x)\lor\lnot At(x,W))\land(S(x)\lor(\lnot Ad(x,y)\lor\lnot At(y,W)))
$$

##### 平滑化
$$
(S(x)\lor\lnot At(x,W))\land(S(x)\lor\lnot Ad(x,y)\lor\lnot At(y,W))
$$

##### 選言の含意への変換
$$
At(x,W)\Rightarrow S(x)\tag{C}
$$
$$
Ad(x,y)\land At(y,W)\Rightarrow S(x)\tag{D}
$$

#### $[2,1]$に隣接する部屋は3つのみ
$$
\forall x(Ad([2,1],x)\Rightarrow x=[1,1]\lor x=[2,2]\lor x=[3,1])
$$

##### 標準化（全称限量子除去）
$$
Ad([2,1],x)\Rightarrow x=[1,1]\lor x=[2,2]\lor x=[3,1]\tag{E}
$$

#### $[1,2]$と$[2,2]$の部屋は隣接
$$
Ad([1,2],[2,2])\tag{F}
$$

### 感覚情報
#### ①,③ $[1,1]$に野獣はいない
$$
\lnot At([1,1],W)
$$

##### 標準化
$$
At([1,1],W)\Rightarrow False\tag{G}
$$

#### ② $[1,2]$で異臭はしない
$$
\lnot S([1,2])
$$

##### 標準化
$$
S([1,2])\Rightarrow False\tag{H}
$$

#### ④ $[2,1]$で異臭がしたが野獣はいない
$$
S([2,1])\land\lnot At([2,1],W)
$$

##### 標準化
$$
S([2,1])\tag{I}
$$
$$
At([2,1],W)\Rightarrow False\tag{J}
$$

### 野獣の位置の推論

#### $[2,2]$に野獣がいないことの証明
(D),(H)において$\{x/[1,2]\}$より
$$
Ad([1,2],y)\land At(y,W)\Rightarrow False\tag{1}
$$

(F),(1)において$\{y/[2,2]\}$より
$$
At([2,2],W)\Rightarrow False\tag{2}
$$

#### $[2,1]$に隣接する部屋のどれかに野獣がいることの証明
(A),(I)において$\{x/[2,1]\}$より
$$
At([2,1],W)\lor Ad([2,1],F([2,1]))\tag{3}
$$

(J),(3)より
$$
Ad([2,1],F([2,1]))\tag{4}
$$

(E),(4)において$\{x/F([2,1])\}$より
$$
F([2,1])=[1,1]\lor F([2,1])=[2,2]\lor F([2,1])=[3,1]\tag{5}
$$

(B),(I)において$\{x/[2,1]\}$より
$$
At([2,1],W)\lor At(F([2,1]),W)\tag{6}
$$

(J),(6)より
$$
At(F([2,1]),W)\tag{7}
$$

(5),(7)より
$$
At([1,1],W)\lor At([2,2],W)\lor At([3,1],W)\tag{8}
$$

#### $[3,1]$に野獣がいることの証明
(G),(8)より
$$
At([2,2],W)\lor At([3,1],W)\tag{9}
$$

(2),(9)より
$$
At([3,1],W)
$$
